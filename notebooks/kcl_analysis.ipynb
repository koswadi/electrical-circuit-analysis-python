# Kirchhoff's Current Law (KCL) Analysis

This notebook demonstrates verification of Kirchhoff's Current Law using Python.

KCL states that the total current entering a node equals the total current leaving the node.

## Formula

Kirchhoff's Current Law:

Σ I_in = Σ I_out

Where:

- I_in = Incoming currents
- I_out = Outgoing currents

Current is conserved at every electrical node.

```python
incoming_currents = [2, 3]
outgoing_currents = [5]

total_in = sum(incoming_currents)
total_out = sum(outgoing_currents)

print(f"Total Incoming Current : {total_in} A")
print(f"Total Outgoing Current : {total_out} A")

if total_in == total_out:
    print("KCL Verified")
else:
    print("KCL Violated")
```

```python
import matplotlib.pyplot as plt

labels = ["Incoming", "Outgoing"]
values = [total_in, total_out]

plt.bar(labels, values)

plt.ylabel("Current (A)")
plt.title("KCL Verification")

plt.show()
```

## KCL Verification Using Dataset

```python
import pandas as pd

df = pd.read_csv("../data/sample_circuit_data.csv")

df.head()
```

```python
df["KCL_Status"] = (
    df["Node_In_A"]
    == df["Node_Out_A"]
)

df[
    [
        "Circuit_ID",
        "Node_In_A",
        "Node_Out_A",
        "KCL_Status"
    ]
]
```

```python
import matplotlib.pyplot as plt

plt.bar(
    df["Circuit_ID"],
    df["Node_In_A"],
    label="Incoming"
)

plt.bar(
    df["Circuit_ID"],
    df["Node_Out_A"],
    alpha=0.5,
    label="Outgoing"
)

plt.legend()

plt.title("KCL Verification")

plt.show()
```

## Conclusion

The node satisfies Kirchhoff's Current Law because the total incoming current equals the total outgoing current.

Current conservation is maintained within the circuit node.
