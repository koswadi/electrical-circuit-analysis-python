# Kirchhoff's Voltage Law (KVL) Analysis

This notebook demonstrates verification of Kirchhoff's Voltage Law using Python.

KVL states that the algebraic sum of all voltages around a closed loop is zero.

## Formula

Kirchhoff's Voltage Law:

Σ V = 0

Where:

- Source voltages are positive
- Voltage drops are negative

The total voltage around a closed loop must equal zero.

```python
source_voltage = 12

voltage_drop_r1 = 4
voltage_drop_r2 = 8

loop_sum = source_voltage - voltage_drop_r1 - voltage_drop_r2

print(f"Loop Voltage Sum = {loop_sum} V")

if loop_sum == 0:
    print("KVL Verified")
else:
    print("KVL Violated")
```

```python
import matplotlib.pyplot as plt

elements = ["Source", "R1 Drop", "R2 Drop"]
voltages = [12, -4, -8]

plt.bar(elements, voltages)

plt.ylabel("Voltage (V)")
plt.title("KVL Verification")

plt.show()
```

## KVL Verification Using Dataset

```python
import pandas as pd

df = pd.read_csv("../data/sample_circuit_data.csv")

df.head()
```

```python
df["KVL_Sum"] = (
    df["Source_Voltage_V"]
    - df["Voltage_Drop_R1_V"]
    - df["Voltage_Drop_R2_V"]
)

df[
    [
        "Circuit_ID",
        "KVL_Sum"
    ]
]
```

```python
import matplotlib.pyplot as plt

plt.plot(
    df["Circuit_ID"],
    df["KVL_Sum"],
    marker="o"
)

plt.axhline(
    y=0,
    linestyle="--"
)

plt.title("KVL Verification")

plt.ylabel("Loop Voltage Sum")

plt.show()
```

## Conclusion

The closed loop satisfies Kirchhoff's Voltage Law.

The source voltage is exactly balanced by the voltage drops across the circuit elements, resulting in a net loop voltage of zero.
