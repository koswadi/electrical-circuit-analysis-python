# Ohm's Law Analysis

This notebook demonstrates the implementation of Ohm's Law using Python.

Ohm's Law describes the relationship between voltage, current, and resistance in an electrical circuit.

It is one of the fundamental principles used in electrical engineering for circuit analysis and design.

## Formula

Ohm's Law:

V = I × R

Where:

- V = Voltage (Volts)
- I = Current (Amperes)
- R = Resistance (Ohms)

Derived forms:

I = V / R

R = V / I

```python
voltage = 12
resistance = 6

current = voltage / resistance

print(f"Voltage    : {voltage} V")
print(f"Resistance : {resistance} Ω")
print(f"Current    : {current:.2f} A")
```

```python
import matplotlib.pyplot as plt

voltages = [5, 10, 15, 20, 25]
resistance = 5

currents = [v / resistance for v in voltages]

plt.plot(voltages, currents, marker='o')
plt.xlabel("Voltage (V)")
plt.ylabel("Current (A)")
plt.title("Ohm's Law: Voltage vs Current")
plt.grid(True)

plt.show()
```
